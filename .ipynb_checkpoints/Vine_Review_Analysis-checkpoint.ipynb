```python
# Import dependencies.
import pandas as pd
```

```python
# Create base dataframe from csv.
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
# Create a new DataFrame  to retrieve all the rows where the total_votes count is equal to or greater than 20.
votes_filter_df = vine_df[vine_df["total_votes"]>=20]
votes_filter_df.head()
```

```python
# Create a new DataFrame to retrieve all the rows where the number of helpful_votes divided 
# by total_votes is equal to or greater than 50%.
secondpass_df = votes_filter_df[(votes_filter_df["helpful_votes"]/votes_filter_df["total_votes"])>=0.5]
secondpass_df.head()
```

```python
# Create a new DataFrame that retrieves all the rows where a review was written as part of the Vine program (paid).
paid_reviews_df = secondpass_df[secondpass_df["vine"]=="Y"]
paid_reviews_df
```

```python
# Create a new DataFrame that retrieves all the rows where a review was not written as part of the Vine program (unpaid).
unpaid_reviews_df = secondpass_df[secondpass_df["vine"]=="N"]
unpaid_reviews_df
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and 
# the percentage of 5-star reviews for paid reviews.
paid_total_reviews = paid_reviews_df["review_id"].count()
paid_total_reviews
```

```python
paid_fivestar_reviews = paid_reviews_df[paid_reviews_df["star_rating"]>=5.0]["review_id"].count()
paid_fivestar_reviews
```

```python
paid_summary = pd.DataFrame({
    "Total Reviews":f'{paid_total_reviews:,}',
    "5-star Reviews": f'{paid_fivestar_reviews:,}',
    "Percentage of 5-star Reviews": f'{float(paid_fivestar_reviews/paid_total_reviews)*100:.2f}%'
}, index=["Paid Reviews"])
paid_summary
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and 
# the percentage of 5-star reviews for unpaid reviews.
total_reviews = unpaid_reviews_df["review_id"].count()
five_star_reviews = unpaid_reviews_df[unpaid_reviews_df["star_rating"]>=5.0]["review_id"].count()
```

```python
unpaid_summary = pd.DataFrame({
    "Total Reviews":f'{total_reviews:,}',
    "5-star Reviews": f'{five_star_reviews:,}',
    "Percentage of 5-star Reviews": f'{float(five_star_reviews/total_reviews)*100:.2f}%'
}, index=["Unpaid Reviews"])
unpaid_summary
```
